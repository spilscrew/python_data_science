## Задание 1

Импортируйте библиотеки pandas и numpy.

Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn. Создайте датафреймы X и Y из этих данных.

Разбейте эти датафреймы н а тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.

Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
```

```python
boston_prices = load_boston()
boston_prices
```

```python
print(boston_prices["DESCR"])
```

```python
feature_names = boston_prices["feature_names"]
feature_names
```

```python
target = boston_prices["target"]
target
```

```python
boston_data = boston_prices["data"]
```

```python
X = pd.DataFrame(boston_data, columns=feature_names)
X.head()
```

```python
Y = pd.DataFrame(target, columns=["price"])
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
Y_pred = lr.predict(X_test)
```

```python
check_test = pd.DataFrame({
    "y_test": Y_test["price"],
    "y_pred": Y_pred.flatten()
})

check_test
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(check_test["y_test"], check_test["y_pred"])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
CREDITCARD_DATASET_PATH = './creditcard.csv'
creditcard_dataset = pd.read_csv(CREDITCARD_DATASET_PATH)
df = pd.DataFrame(creditcard_dataset)
```

```python
df["Class"].value_counts(normalize=True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
X = pd.DataFrame(df).drop(["Class"], axis=1)
X
```

```python
y = pd.Series(df["Class"])
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
parameters = [{'n_estimators': [10, 15], 'max_features': np.arange(3, 5), 'max_depth': np.arange(4, 7)}]
```

```python
clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_pred_proba = clf.predict_proba(X_test)[:, 1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
